```python
import pandas as pd
import matplotlib.pyplot as plt


df= pd.read_excel('Online Retail.xlsx')
df
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract month and day of the week from InvoiceDate
df['Month'] = df['InvoiceDate'].dt.month
df['DayOfWeek'] = df['InvoiceDate'].dt.dayofweek

# Visualize sales trends over months
plt.figure(figsize=(10, 6))
df.groupby('Month')['Quantity'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Sales per Month')
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=0)
plt.show()

# Visualize sales trends over days of the week
plt.figure(figsize=(10, 6))
df.groupby('DayOfWeek')['Quantity'].sum().plot(kind='bar', color='salmon')
plt.title('Total Sales per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=0)
plt.show()

# Visualize the distribution of Quantity using a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Quantity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Visualize the distribution of UnitPrice using a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df['UnitPrice'], vert=False)
plt.title('Distribution of Unit Price')
plt.xlabel('Unit Price')
plt.show()

# Visualize the frequency of countries using a bar plot
plt.figure(figsize=(10, 6))
df['Country'].value_counts().plot(kind='bar', color='salmon')
plt.title('Frequency of Countries')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()
```
